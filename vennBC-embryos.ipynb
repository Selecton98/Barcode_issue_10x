```r
sample110x<-read.csv('rep1BC_embryo.csv')
sample1clone<-read.csv(file='Rep1-AlleleColonies.txt',sep=',',header=F)
sample1clone<-as.data.frame(t(as.matrix(sample1clone)))
clonebc<-sample1clone[,2]
for (i in 3:dim(sample1clone)[2]){
tmp<-sample1clone[,i]
clonebc<-c(clonebc,tmp)
}
clonebc<-unique(clonebc)
clonebc<-clonebc[-18]
```

```r
library(VennDiagram)
options(repr.plot.height=5,repr.plot.width=5)
input<-list(sample110x$bc,clonebc,sample1clone[,1])
Table<-calculate.overlap(input)
venn<-venn.diagram(input,NULL, cat.fontface = "bold",cat.cex = 1,cex = 1,
category = c('BC-transcriptome','BC-edited',"BC-unedited"),fill = c( "lightgrey","firebrick",'dodgerblue'),lty='blank',
cat.col= 'black',   ) 
grid.draw(venn)
```
